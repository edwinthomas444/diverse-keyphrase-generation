```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()


dataset = "kp20k"
all_pred_path = f'/home/thomased/work/codebase/coopsummer2023/Project/checkpoints/phase2_exp/abstractive_models/precision_heads_test2_final/{dataset}_predictions.txt'
prec_pred_path = f'/home/thomased/work/codebase/coopsummer2023/Project/checkpoints/phase2_exp/abstractive_models/precision_heads_test2_final/{dataset}_predictions_7.txt'

def stem_norm(container):
    result_set = set()
    remove_space_set = set([w.strip() for w in container])
    for keyphrase in remove_space_set:
        stem_keyphrase_list = []
        for word in keyphrase.split(" "):
            stem_keyphrase_list.append(stemmer.stem(word))
        result_set.add(" ".join(stem_keyphrase_list))
    return result_set



with open(all_pred_path, 'r') as f1, open(prec_pred_path, 'r') as f2:
    lines = f1.read().split('\n\n')
    lines1 = f2.read().split('\n\n')
    
    
    for i, (line, line1) in enumerate(zip(lines, lines1)):
        gt = line.split('\n')[-2].strip().replace('Ground Truth: ', '')
        pred = line.split('\n')[-1].strip().replace('Pred Text: ', '')
        gt1 = line1.split('\n')[-2].strip().replace('Ground Truth: ', '')
        pred1 = line1.split('\n')[-1].strip().replace('Pred Text: ', '')
        
        gts = set(gt.split(' ; '))
        preds = set(pred.split(' ; '))
        # gts1 = set(gt1.split(' ; '))
        preds1 = set(pred1.split(' ; '))
        
        # if there are intersections 
        # find ones which were missed
        missed = stem_norm(gts)&stem_norm(preds) - stem_norm(preds1)
        if missed:
            print('missed in precision heads results: ',missed)
            print('all precision head results: ',preds1)
            print('\n')
        
        
        
            
```
