```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import numpy as np



# Assuming you have paragraph embeddings stored in 'embeddings' variable of shape (384, 768)

def plot(embeddings):
    # Initialize t-SNE
    tsne = TSNE(n_components=3, random_state=42, perplexity=1)

    # Fit and transform the embeddings using t-SNE
    embeddings_2d = tsne.fit_transform(embeddings)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    # Plot the embeddings
    for i in range(len(embeddings)):
        # plt.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], label=f'Tensor {i+1}', marker='o')
        ax.scatter(embeddings_2d[i, 0], embeddings_2d[i, 1], embeddings_2d[i, 2], marker='o')
        ax.text(embeddings_2d[i, 0], embeddings_2d[i, 1], embeddings_2d[i, 2], f'Tensor {i+1}', verticalalignment='bottom', horizontalalignment='right')

    # plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])
    ax.set_title("Paragraph Embeddings Visualization")
    ax.set_xlabel("PC-1")
    ax.set_ylabel("PC-2")
    ax.set_zlabel("PC-3")
    # plt.legend()
    plt.show()


def kmeans(embeddings):
    # Generate random embeddings data
    # num_embeddings= lesn(embeddings)
    # embedding_size = 768
    # embeddings = np.random.randn(num_embeddings, embedding_size)

    # Perform K-means clustering
    num_clusters = 2
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    labels = kmeans.fit_predict(embeddings)
    print('labels: ', labels)
    return labels

def accuracy(pred_labels, dhead_gt):
    import numpy as np
    # count = 0
    # count_total = 0
    # for x, y in zip(pred_labels, dhead_gt):
    #     if y == 1:
    #         count_total+=1
    #     if x == y and y==1:
    #         count+=1
    #Get the confusion matrix
    cm = confusion_matrix(dhead_gt, pred_labels)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    return cm

for ind, d in enumerate(data):
    # print(d.keys())
    dhead_gt = d['dhead_gt']
    dhead_pred = d['dhead_kp']
    dhead_key = d['dhead_key'].to('cpu')
    dhead_val = d['dhead_val'].to('cpu')
    if 1 in dhead_gt:
        # print(dhead_gt)
        print(dhead_pred)
        dhead_key = dhead_key.mean(dim=1)
        dhead_val = dhead_val.mean(dim=1)
        print(dhead_key.shape)
        combined = torch.matmul(dhead_key, dhead_val.permute(1,0))
        print(combined.shape)
        plot(dhead_val)
        key_labels = kmeans(dhead_key)
        val_labels = kmeans(dhead_val)
        pred_labels = kmeans(combined)
        
        kacc = accuracy(key_labels, dhead_gt)
        vacc = accuracy(val_labels, dhead_gt)
        cacc = accuracy(pred_labels, dhead_gt)
        print('accuracy: ', kacc.diagonal(), vacc.diagonal(), cacc.diagonal())
        print('gt labels: ', dhead_gt)
        print('\n')
    
    # if ind>100:
    #     break
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids

# Generate random embeddings data
num_embeddings = 8
embedding_size = 768
embeddings = np.random.randn(num_embeddings, embedding_size)

# Set the number of clusters (K)
num_clusters = 2

# Perform K-medoids clustering
kmedoids_instance = KMedoids(n_clusters=num_clusters, random_state=42)
clusters = kmedoids_instance.fit_predict(embeddings)

# Print cluster assignments
for i in range(num_clusters):
    cluster = np.where(clusters == i)[0]
    print(f"Cluster {i+1}: {cluster}")

# Print medoids
medoids = kmedoids_instance.medoid_indices_
print("Medoids:", medoids)
```
